```python
# get all header
def get_headers(df):
    headers = list(df.columns.values)
    return headers

def feature_value_counts(df, detail = True):
    features = get_headers(df)
    for feature in features:
        value = df[feature].value_counts()
        if detail:
            print(feature)
            print(value)
        else:
            print(feature, len(value))
```
